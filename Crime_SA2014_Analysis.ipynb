```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cs = pd.read_csv('Crime_Statistics_SA_2014_2019.csv')
```

```python
#A2
bool_series = pd.isnull(cs["Reported Date"])|pd.isnull(cs["Suburb - Incident"])|pd.isnull(cs["Postcode - Incident"])|pd.isnull(cs["Offence Level 1 Description"])|pd.isnull(cs["Offence Level 2 Description"])|pd.isnull(cs["Offence Level 3 Description"])|pd.isnull(cs["Offence Count"])
cs[bool_series]
```

```python
#A3
print("Data types before change",cs.dtypes)
cs['Reported Date'] = cs['Reported Date'].astype("datetime64")
print("\nMin is",cs['Reported Date'].min())
print("\nMax is",cs['Reported Date'].max())
print("\nData types after change",cs.dtypes)
```

```python
#B1
cs['year']=cs['Reported Date'].dt.year
timeseries = cs.groupby('year').count()
timeseries = timeseries.drop(['Suburb - Incident','Reported Date','Offence Level 3 Description','Postcode - Incident','Offence Level 1 Description','Offence Level 2 Description'],axis = 1)
timeseries = timeseries.reset_index() 
timegraph = timeseries.plot(x = 'year',y = 'Offence Count',color = 'orange',legend=False)
plt.title("Number of Crimes in South Australia")
plt.ylabel("Offence Count")
plt.xlabel("Year")
```

```python
#B2
groupbySuburbs = cs.groupby('Suburb - Incident').sum() #.agg({'Offence Count':'sum'})
groupbySuburbs = groupbySuburbs.drop(['year'],axis = 1)
print(groupbySuburbs.mean(),groupbySuburbs.median())
hist_plot = groupbySuburbs.hist(bins = 50)
plt.title("Frequence of Offences in Different Suburbs")
plt.ylabel("Number of suburbs")
plt.xlabel("Offence Count")
```

```python
filt = groupbySuburbs['Offence Count']>5000 
dangerous_Suburbs = groupbySuburbs[filt]
ax = dangerous_Suburbs.plot.barh(rot=0,width = 0.7,color = "orange")
plt.title("Total Number of Crimes in Dangerous Suburbs")
plt.xlabel("Number of crimes")
```

```python
#B3
groupbyDateSuburb = cs.groupby(['Reported Date','Suburb - Incident']).sum()
groupbyDateSuburb = groupbyDateSuburb.drop(['year'],axis = 1)
groupbyDateSuburb = groupbyDateSuburb.reset_index() 
filt = groupbyDateSuburb['Offence Count']>15
HighIncidentSuburbs = groupbyDateSuburb[filt]
HighIncidentSuburbs
```

```python
HighIncidentSuburbsFreq = HighIncidentSuburbs.groupby('Suburb - Incident').count()
HighIncidentSuburbsFreq = HighIncidentSuburbsFreq.drop(['Reported Date'],axis = 1)
HighIncidentSuburbsFreq = HighIncidentSuburbsFreq.reset_index() 
HighIncidentSuburbsFreq
```

```python
HighIncidentSuburbsFreq.plot.bar(x = 'Suburb - Incident',width = 0.7,color = "orange",legend = False)
plt.title("Frequency of High Offence Count Days in Different Suburbs")
plt.ylabel("Frequency")
```

```python
HighIncidentSuburbsFreq.boxplot(column = 'Offence Count')
plt.title("Boxlplot of High Offence Count Days in Different Suburbs")
plt.ylabel("Frequency")
```

```python
noOutlier = HighIncidentSuburbsFreq['Suburb - Incident'] != 'ADELAIDE'
noOutlierGroup = HighIncidentSuburbsFreq[noOutlier]
noOutlierGroup.plot.bar(x = 'Suburb - Incident',width = 0.7,color = "orange",legend = False)
plt.title("Frequency of High Offence Count Days in Different Suburbs")
plt.ylabel("Frequency")
```
